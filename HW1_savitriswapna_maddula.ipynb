### Can you write a function to return coefficients based on stochastic gradient descent method 

```python
dataset = [[1, 1], [2, 3], [4, 3], [3, 2], [5, 5], [6, 6]]
l_rate = 0.01
n_epoch = 50

def calculate_coeff(dataset, l_rate, n_epoch):
    coef = [0.0,0.0]
    for n in range(n_epoch):
        error = 0
        for entry in dataset:
            predicted_value = coef[0] + (coef[1]*entry[0])
            error = predicted_value - entry[1]
            coef[0] = coef[0] - (l_rate*error)
            coef[1] = coef[1] - (l_rate*error*entry[0])
    return coef

coef = calculate_coeff(dataset, l_rate, n_epoch)
print("The cofficients calculated by stochastic gradient descent method are",coef)
```

```python
def bernouli():
    bernouli_nb=BernoulliNB(binarize=.08)
    bernouli_nb.fit(mini_train_data, mini_train_labels)
    prediction=bernouli_nb.predict(dev_data)
    matching=prediction==dev_labels
    accuracy=float(len([match for match in matching if match==True]))/float(len(matching))
    print("Accuracy in BernouliNB: " + str(accuracy))

def multinomial():
    pixlated_train_data = mini_train_data
    pixlated_dev_data = dev_data
    #map training data
    for i in range(len(pixlated_train_data)):
        for j in range(len(pixlated_train_data[0])):
            if(pixlated_train_data[i][j]<0.33):
                pixlated_train_data[i][j] = 0
            elif(pixlated_train_data[i][j]>0.66):
                pixlated_train_data[i][j] = 2
            else:
                pixlated_train_data[i][j] = 1
        
    #map dev data
    for i in range(len(pixlated_dev_data)):
        for j in range(len(pixlated_dev_data[0])):
            if(pixlated_dev_data[i][j]<0.33):
                pixlated_dev_data[i][j] = 0
            elif(pixlated_dev_data[i][j]>0.66):
                pixlated_dev_data[i][j] = 2
            else:
                pixlated_dev_data[i][j] = 1  
    
    multinomial_nb=MultinomialNB()
    multinomial_nb.fit(pixlated_train_data, mini_train_labels)
    prediction=multinomial_nb.predict(pixlated_dev_data)
    matching=prediction==dev_labels
    accuracy=float(len([match for match in matching if match==True]))/float(len(matching))
    print("Accuracy in MultinomialNB: " + str(accuracy))

bernouli()
multinomial()
```

There is no accuracy improvement in multinomial case. 
I have chosen threshold values as 0.33 and 0.66 for mapping data as 0,1,2.
This may increase similarity between many digits as a group of them are mapped to a single value. 
This may be the reason in reduction of accuracy.
